### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

def validate_data(x, y):
    if not isinstance(x, np.ndarray) or not isinstance(y, np.ndarray):
        raise TypeError("Input data must be NumPy arrays.")
    if x.shape != y.shape:
        raise ValueError("Input arrays must have the same shape.")
    if x.size == 0 or y.size == 0:
        raise ValueError("Input arrays must not be empty.")
    if np.any(np.isnan(x)) or np.any(np.isnan(y)):
        raise ValueError("Input arrays must not contain NaN values.")
    if np.any(np.isinf(x)) or np.any(np.isinf(y)):
        raise ValueError("Input arrays must not contain infinite values.")

try:
    x = np.linspace(0, 10, 100)
    y = x ** 2
    
    validate_data(x, y)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=r'$y=x^2$', color='b')
    plt.title("Line Plot")
    plt.xlabel("X Values")
    plt.ylabel("Y Values")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()
except Exception as e:
    print(f"Error: {e}")
```
