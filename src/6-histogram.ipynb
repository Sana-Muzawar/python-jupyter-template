### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt

def validate_data_histogram(data):
    if not isinstance(data, list):
        raise TypeError("Data must be a list.")
    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError("Data must contain only numeric values.")
    if len(data) == 0:
        raise ValueError("Data list must not be empty.")

try:
    data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
    
    validate_data_histogram(data)
    
    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=5, color='green', edgecolor='black')
    plt.title("Histogram")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()
except Exception as e:
    print(f"Error: {e}")
```
